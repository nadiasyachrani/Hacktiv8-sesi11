### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels
import sklearn
%matplotlib inline
```

### Simple Linear Regression With scikit-learn

```python
x = np.array([5, 15, 25, 35, 45, 55])
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
# jadi 2D =>> x harus 2D
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
```

```python
print(x)
print(y)
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
# R square makin mendekati 1 makin bagus
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
```

```python
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

* y = mx + b
* y adalah target
* m adalah slope
* b adalah intercept

```python
print('Formula dari model adalah y = ',model.coef_[0],'x+',model.intercept_)
```

```python
# memprediksi model
y_pred = model.predict(x)
y_pred
```

```python
print(x)
print(y)
```

```python
# bandingkan y dan y_pred
plt.scatter(x, y, alpha=0.5)
plt.plot(x, y_pred)

plt.title('Scatter plot x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_new = np.arange(5,20,5).reshape((-1, 1))
print(x_new)
```

```python
y_new_pred = model.predict(x_new)
print(y_new_pred)
```

### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([15, 11, 2, 8, 25, 32])
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y)
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
y_pred = model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
model.score(x,y)
```

```python
# mengubah feature menjadi x^2, model mendekati scatter plotnya
```

```python
transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(x)
```

```python
x_poly = transformer.transform(x)
x_poly
```

```python
model_poly = LinearRegression()
model_poly.fit(x_poly,y)
```

```python
model_poly.score(x_poly,y)
```

```python
model.score(x,y)
```

```python
y_poly_pred = model_poly.predict(x_poly)
y_poly_pred
```

```python
plt.scatter(x,y,color='green',label='true y')
plt.plot(x,y_poly_pred,color='red',label='predicted y')
plt.legend()
plt.show()
```

```python
x = np.array([[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]])
y = np.array([4, 5, 20, 14, 32, 22, 38, 43])
```

```python
x_poly = transformer.fit_transform(x)
model_poly = LinearRegression()
model = LinearRegression()
model_poly.fit(x_poly,y)
model.fit(x,y)
```

```python
print(model.score(x,y),model_poly.score(x_poly,y))
```

### Advanced Linear Regression With statsmodels

```python
import statsmodels.api as sm
#skilearn lebih banyak model mechine learning
#statmodels lebih banyak algoritma, bisa lihat summary dari modelnya sendiri
```

```python
x = np.array([[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]])
y = np.array([4, 5, 20, 14, 32, 22, 38, 43])
```

```python
x = sm.add_constant(x)
```

```python
x
```

```python
#OLS regression
model_stat = sm.OLS(y,x)
```

```python
results = model_stat.fit()
```

```python
print(results.summary())
```

formula regresi y = 5.5226 + 0.4471x1 + 0.2550x2

```python
print('rsquared:',results.rsquared)
```

```python
results.params
```

```python
res_summary = results.summary()
res_summary.tables[1]
```

```python
res_summary.tables[0]
```

```python
y_pred = results.predict(x)
y_pred
```

```python
mean_squared_error(y,y_pred)
```

### Contoh 2 sklearn

```python
df1 = pd.read_csv('CarPrice_Assignment.csv',usecols=['horsepower','price'])
df1
```

```python
f = plt.figure(figsize=(12,4))

f.add_subplot(1,2,1)
df1['horsepower'].plot(kind='kde')

f.add_subplot(1,2,2)
df1['horsepower'].plot(kind='box')

plt.show()
```

```python
X = df1['horsepower']
y = df1['price']
plt.scatter(X,y)
plt.ylabel('price')
plt.xlabel('horsepower')
plt.show()
```

```python
df1.corr()
```

```python
X = X.to_frame()
```

## Data Splitting


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
linreg.score(X_train,y_train)
```

```python
y_pred = linreg.predict(X_test)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred)
```

```python
y_pred_train = linreg.predict(X_train)
y_pred_test = linreg.predict(X_test)
```

```python
mse_train = mean_squared_error(y_train,y_pred_train)
mse_test = mean_squared_error(y_test,y_pred_test)

print('train mse:', mse_train)
print('test_mse:', mse_test)
```

```python
rmse_tr = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)

print('train rmse:', rmse_tr)
print('test rmse:', rmse_test)
```
